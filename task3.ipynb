# Use a grouped bar chart to compare the average rating and total review count for the top 10 app categories by number of installs. Filter out any categories where the average rating is below 4.0 and size below 10 M and last update should be Jan month . this graph should work only between 3PM IST to 5 PM IST apart from that time we should not show this graph in dashboard itself.


```python
import pandas as pd
import numpy as np
from datetime import datetime
import plotly.graph_objects as go
import pytz
```

# load dataset

```python
apps = pd.read_csv('google playstore data.csv')
# print(apps.to_string())
apps.head()
```

```python
# dataclean 
apps = apps[apps['Installs'].str.contains(r'\d',regex=True,na=False)]
apps['Installs'] = apps['Installs'].str.replace('[,+]','',regex=True).astype(float)
# apps.head()
apps = apps[apps['Size'].str.contains(r'\d',regex=True,na=False)]
apps['Size']= apps['Size'].str.replace('M','',regex=True)
apps['Size']= apps['Size'].str.replace('k','',regex=True).astype(float)/1025
apps.head()


```

```python
filter_df = apps[(apps['Rating']>=4.0) & (apps['Size']>=0.01)]
filter_df.head()
```

```python
filter_df['Last Updated'] = pd.to_datetime(filter_df['Last Updated'])
filter_df.head()
filter_df= filter_df[filter_df['Last Updated'].dt.month==1]
# filter_df.head()
```

```python
top10_cat = filter_df.groupby('Category').agg({ 'Rating':'mean','Reviews':'count','Installs':'sum'})
top10_cat = top10_cat.nlargest(10,'Installs').reset_index()
# top10_groups=top10_cat['Category'].tolist()
top10_cat
# top10_groups
```

```python
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist)
current_hour = current_time.hour

if 15 <= current_hour < 17:  # Correct time range for 3 PM to 5 PM IST
    fig = go.Figure()

    
    fig.add_trace(go.Bar(
        name='Average Rating',
        x=top10_cat['Category'],
        y=top10_cat['Rating'],
        yaxis='y1'
    ))

    
    fig.add_trace(go.Bar(
        name='Total Reviews',
        x=top10_cat['Category'],
        y=top10_cat['Reviews'],
        yaxis='y2'
    ))

    fig.update_layout(
        title='Average Rating and Total Reviews for Top 10 App Categories',
        xaxis=dict(title='App Categories'),
        yaxis=dict(title='Average Rating', side='left'),
        yaxis2=dict(
            title='Total Reviews',
            overlaying='y',
            side='right'
        ),
        barmode='group'
    )

    fig.show()
else:
    print("Graph can only be viewed between 3 PM and 5 PM IST.")
```
