# Create a violin plot to visualize the distribution of ratings for each app category, but only include categories with more than 50 apps and app name should contain letter “C” and exclude apps with fewer than 10 reviews and rating should be less 4.0. this graph should work only between 4 PM IST to 6 PM IST apart from that time we should not show this graph in dashboard itself.


```python
import pandas as pd
import plotly.express as px
import datetime as dt
import pytz

```

```python
app = pd.read_csv('google playstore data.csv')
app.head()
```

```python
app.info()
```

```python
app = app[app['Reviews'].str.contains(r'\d',regex=True,na=False)]
app['Reviews']=pd.to_numeric(app['Reviews'],errors='coerce')
app['Reviews']=app['Reviews'].fillna(0).astype(int)
app['Reviews'].info()
```

```python
app['Rating'] = app['Rating'].fillna(app['Rating'].mean())
filter_app = app[(app['Rating'] < 4.0) & (app['Reviews']>10)&(app['App'].str.contains('C',case=False,na=False))]
filter_app.head()
```

```python
category_count=filter_app['Category'].value_counts()
valid_Category=category_count[category_count>50].index
valid_Category
filter_app=filter_app[filter_app['Category'].isin(valid_Category)]
filter_app
```

```python
IST = pytz.timezone('Asia/Kolkata')
current_time = dt.datetime.now(IST).time()
start_time = dt.time(16,0)
end_time = dt.time(18,0)

if start_time <= current_time <= end_time:
    # Create Violin Plot using Plotly
    fig = px.violin(filter_app, x='Category', y='Rating', box=True, points="all",
                    title="Distribution of Ratings for Each App Category",
                    labels={"Category": "App Category", "Rating": "Rating"})
    
    fig.update_layout(xaxis_tickangle=-45)  # Rotate category labels for readability
    fig.show()
else:
    print("Graph is only visible between 4 PM - 6 PM IST.")
```
